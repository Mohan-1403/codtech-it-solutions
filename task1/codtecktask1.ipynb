```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .master("local[*]") \
    .appName("CODTECH Big Data Analysis") \
    .getOrCreate()

print("Spark Started")
```

```python
df = spark.read.json("Cell_Phones_and_Accessories_5.json")
df.show(5)
```

```python
from pyspark.sql.functions import col, avg, length

df_clean = df.dropna(subset=["overall", "reviewText"])

df_clean.groupBy("overall").count().show()

df_clean.select(avg("overall")).show()

df_clean.groupBy("asin").count().orderBy(col("count").desc()).show(10)

df_clean = df_clean.withColumn("review_length", length(col("reviewText")))
df_clean.select(avg("review_length")).show()
```
